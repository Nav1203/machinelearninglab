```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
```

```python
train_data=pd.read_csv("train (1).csv") 
```

```python
train_data.head()
```

```python
le=LabelEncoder()
for i in ['Location','Urban','US']:
  train_data[i]=le.fit_transform(train_data[i])
```

```python
train_data.isnull().sum()
```

```python
min(train_data['Sales'])
```

```python
train_data.head()
```

```python
y=pd.cut(train_data.Sales,bins=[0,5,10,50],labels=[1,2,3],include_lowest=True)
```

```python
x=train_data.drop(['Sales','ID'],axis=1)
```

```python
display(x)
```

```python
display(y)
```

```python
x.isnull().sum()
```

```python
model=SVC(kernel='linear')
```

```python
model.fit(x,y)
```

```python
test_data=pd.read_csv("test (1).csv")
```

```python
for i in ['Location','Urban','US']:
  test_data[i]=le.fit_transform(test_data[i])
```

```python
test_data.head()
```

```python
x_test=test_data.drop(['Sales','ID'],axis=1)
```

```python
x_test.head()
```

```python
y_test=pd.cut(test_data.Sales,bins=[0,5,10,50],labels=[1,2,3],include_lowest=True)
```

```python
y_test
```

```python
ypred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
print(classification_report(ypred,y_test))
```

```python
print(accuracy_score(ypred,y_test))
```

```python
print(confusion_matrix(ypred,y_test))
```
